```python
import sympy as sp
import numpy as np
from sympy.physics.vector import dynamicsymbols as dynamicsymbols
```

```python
m_1, m_2, l_1, l_2, g, t= sp.symbols('m_1 m_2 l_1 l_2 g t')
u = sp.symbols('u')
theta_1, theta_2 = dynamicsymbols('theta_1 theta_2')
theta_1_dot = sp.diff(theta_1, t)
theta_2_dot = sp.diff(theta_2, t)
```

```python
state = sp.Matrix([theta_1, theta_1_dot, theta_2, theta_2_dot])

input = sp.Matrix([u])

dynamics = sp.Matrix([theta_1_dot,
                     -(m_2 * (u + l_1 * theta_1_dot ** 2 * sp.sin(theta_1 - theta_2) - g * sp.sin(theta_2)) + m_2 * l_2 * theta_2_dot ** 2 * sp.sin(theta_1 - theta_2) + (m_1 + m_2) * g * sp.sin(theta_1)) / (l_1 * (m_1 + m_2 - m_2 * sp.cos(theta_1 - theta_2))),
                      theta_2_dot,
                      sp.cos(theta_1 - theta_2) * (m_2 * (u + l_1 * theta_1_dot ** 2 * sp.sin(theta_1 - theta_2) - g * sp.sin(theta_2)) + m_2 * l_2 * theta_2_dot ** 2 * sp.sin(theta_1 - theta_2) + (m_1 + m_2) * g * sp.sin(theta_1)) / (l_2 * (m_1 + m_2 - m_2 * sp.cos(theta_1 - theta_2))) + (u + l_1 * theta_1_dot ** 2 * sp.sin(theta_1 - theta_2) - g * sp.sin(theta_2)) / l_2])

```

```python
A = dynamics.jacobian(state)
B = dynamics.jacobian(input)
```

```python
A.subs([
    (theta_1, 0),
    (theta_2, 0),
    (theta_1_dot, 0),
    (theta_2_dot, 0),
    (u, 0)
])
```

```python
B.subs([
    (theta_1, 0),
    (theta_2, 0),
    (theta_1_dot, 0),
    (theta_2_dot, 0),
    (u, 0)
])
```
